## 2. Data Cleaning
'We handle missing values, drop irrelevant columns, and parse date information.'

```python
# Displaying the total number of missing values in each column of the DataFrame
df.isnull().sum()
```

```python
# Calculating the percentage of missing values per column
null_column_percentage = df.isnull().sum()/df.shape[0]*100

# Displaying the percentage of missing values for each column
null_column_percentage
```

```python
# Filling missing values in the 'director' column with 'Unknown' because director information is often missing and replacing it with a placeholder is better than dropping rows or using imputation methods that might not be suitable for categorical data.
df['director'].fillna('Unknown', inplace=True)

# Filling missing values in the 'cast' column with 'Not Mentioned' for the same reasons as the 'director' column.
df['cast'].fillna('Not Mentioned', inplace=True)

# Filling missing values in the 'country' column with 'Unknown' as country information is frequently missing and a placeholder maintains row count while indicating missingness.
df['country'].fillna('Unknown', inplace=True)

# Filling missing values in the 'rating' column with 'Not Rated' as rating information is sometimes unavailable, and a placeholder prevents data loss.
df['rating'].fillna('Not Rated', inplace=True)

# Displaying the first 25 rows to confirm missing values were filled
display(df.head(25))
```

```python
#Before
# Dropping the 'description' column as it will not be used in this analysis
df.drop(columns=['description'], inplace=True)
```

```python
# Displaying the first 10 rows after dropping the 'description' column
df.head(10)
```

```python
# Converting the 'date_added' column to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce', format='mixed')
```

```python
# Extracting 'year_added' and 'month_added' columns from 'date_added'
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()
```

```python
df.head()
```

```python
df.head()
```

```python
# checking if any null values in the Database
df.isnull().sum()
```

```python
# Dropping the subset 'date_added' rootcause for null data in Year_added and Month_added
df = df.dropna(subset=['date_added'])
# Dropping the subset 'Duration' which have null value
df = df.dropna(subset=['duration'])
```

```python
# Changing the data type of 'year_added' from 'float' to 'integer'
df['year_added'] = df['year_added'].astype(int)
df.info()
```

```python
# Creating a heatmap to visualize the missing values in the DataFrame 'df_1' using seaborn
sns.heatmap(df.isnull(), cbar=False)
plt.title('Missing Values After Dropping Rows')
plt.show()
```

```python
# checking the duplicates in the dataframe
df.duplicated().sum()
```

## 3. Exploratory Data Analysis

'Below, we analyze the dataset through descriptive statistics and visualizations.'

```python
# Visualizing the type of content Netflix focuses on through countplot visualization
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Number of Titles')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()

# Value counts
df['type'].value_counts()
```

```python
#Insight
"Netflix focuses more on Movies than TV Shows"
```

```python
df_cast = df[df['cast'] != 'Not Mentioned'].copy()
```

```python
# Creating a list of actors per row
df_cast['cast_list'] = df_cast['cast'].str.split(', ')
df_cast_exploded = df_cast.explode('cast_list')
```

```python
# Counting top 15 actors appreance
actor_counts = df_cast_exploded['cast_list'].value_counts().head(15)
```

```python
#Visualizing the top 15 Frequent Actors on Netflix
plt.figure(figsize=(10,6))
sns.barplot(y=actor_counts.index, x=actor_counts.values, palette='coolwarm')
plt.title('Top 15 Most Frequent Actors on Netflix')
plt.xlabel('Number of Appearances')
plt.ylabel('Actor')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
# Which are the top 10 countries with the  more titles
top_countries = df['country'].value_counts().head(10)
print(top_countries)

# Visualzing the top 10 countries
plt.figure(figsize=(10,6))
sns.barplot(y=top_countries.index, x=top_countries.values, palette="Set3")
plt.title("Top 10 Countries Producing Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
# Insights
"USA dominates in both Movies & TV Shows by producing more than 2500+ titles"
```

```python
#Visualizing the content addition of Netflix yearwise
plt.figure(figsize=(12,5))
df['year_added'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Content Added Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
#Insight
"Most of the content is added after 2016 and peaked in 2019-20 due to pandemic and global expansion"
```

```python
#Visualizing the content addition of Netflix monthwise and also the seasonal trends
plt.figure(figsize=(10,5))
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
df['month_added'].value_counts().loc[month_order].plot(kind='bar', color='orange')
plt.title('Content Added by Month')
plt.xlabel('Month')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
# Insights
"Highest content uploads happen in July and December, hinting at summer vacation and holiday strategies"
```

```python
# Analyzing the types of content based on age groups
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='rating', hue='type', order=df['rating'].value_counts().index, palette='Set2')
plt.title('Rating-wise Content Split by Type')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
#insights
"TV-MA and TV-14 ratings have the highest counts for both Movies and TV Shows, indicating a focus on mature audiences."
```

```python
# Analyzing Are the movies longer than shows of Netflix
plt.figure(figsize=(10,5))

# For movies, the duration is in minutes, so we can extract the numerical part
movie_durations = df[df['type'] == 'Movie']['duration'].str.replace(' min', '').astype(float)

sns.boxplot(x=df[df['type'] == 'Movie']['type'], y=movie_durations, palette='cool')
plt.title('Duration Distribution of Movies')
plt.ylabel('Duration (minutes)')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
#Insights
"Most movies lies between 80 to 120 minutes"
```

```python
# Analyzing which country produces more movie or shows on netflix
top_countries = df['country'].value_counts().head(10).index
plt.figure(figsize=(10,6))
sns.countplot(y='country', hue='type', data=df[df['country'].isin(top_countries)], palette='Spectral')
plt.title('Top Countries Producing Content by Type')
plt.xlabel('Count')
plt.ylabel('Country')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
#Insights
"USA dominates in both Movies & TV Shows. India contributes significantly to Movies"
```
